```python
def get_article(url):  
    try:
        article = Article(url['Sources 1'])
        article.download()
        article.parse()
        article_text = article.text
    
    except:
        article_text = "download failed"
    
    if len(article_text) < 100:
        try:
            article = Article(url['Sources 2'])
            article.download()
            article.parse()
            article_text = article.text
        except:
            article_text = "download failed"
    
    if len(article_text) < 100:
        try:
            article = Article(url['Sources 3'])
            article.download()
            article.parse()
            article_text = article.text
        except:
            article_text = "download failed"
    
    
    
    
    return article_text
    
```

```python
df['text'] = df.apply(lambda x: get_article(x), axis=1)
```
