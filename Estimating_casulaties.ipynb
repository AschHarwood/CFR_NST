```python
df = pd.read_pickle('Complete_NST_Text.pkl')

#cleans \n slashes from text
df['text'] = df['text'].replace(r'\n', ' ', regex=True)

#creates a list of states
state_list = list(df['State'].groupby(df['State']).count().index)

#applies a number to each state
state_dict = {}
for idx, state in enumerate(state_list):
    state_dict[state] = idx

#drops any rows where State = NA
df = df.dropna(subset=['State'])

#applies the code to each row
df['state_code'] = df['State'].apply(lambda x: state_dict[x])
```

```python
def grab_words(text):

    #stop_words = set(stopwords.words('english'))
    text = re.sub(r'\W', " ", text)
 
    text = word_tokenize(text)
    #filtered_sentence = [w.lower() for w in text if not w in stop_words]
    words_to_check = ['killed', 'dead', 'died']
    word_list = []
    for word_1 in words_to_check: #iterates through the words to check
        for key, word_2 in enumerate(text): #iterates through the text
            if word_2 == word_1: #checks if each word in text matches key word
                word_list.append(text[key-5:key+5]) #appends word and surrounding words

    flat_list = [item for sublist in word_list for item in sublist] 
    final_string = ' '.join(flat_list)
    if len(final_string)>0:
        return final_string
    else:
        return "none"
    
    
```

```python
df['kill_words'] = df['text'].apply(lambda x: grab_words(x))
```

```python
df.columns
```

```python
def cat_killed(x):


    if int(x) == 0:
        return 0
    elif int(x) > 0 and int(x) < 6:
        return 1
    elif int(x) > 5 and int(x) < 11:
        return 2
    elif int(x) > 10 and int(x) < 16:
        return 3
    elif int(x) > 15 and int(x) < 21:
        return 4
    elif int(x) > 20:
        return 5
    else:
        return 'not a digit'

    


        
```

```python
df['cat_killed'] = df['Total Deaths'].apply(lambda x: cat_killed(x))
```

```python
df['TD_Types'] = df['Total Deaths'].apply(lambda x: type(x))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import KFold
```

```python
vect = CountVectorizer(stop_words='english', ngram_range=(1,4)).fit(X_train)
```

```python
X_train_vectorized = vect.transform(X_train)
```

```python
model.fit(X_train_vectorized, y_train)
```

```python
predictions = model.predict(vect.transform(X_test))
accuracy_score(y_test, predictions)
```
